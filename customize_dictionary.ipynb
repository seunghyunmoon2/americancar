```python
#from konlpy.tag import Okt
# okt = Okt()
from ckonlpy.tag import Twitter
twitter = Twitter()

stopwords = ['의', '가', '이', '은', '들', '는', '좀', '잘', '걍', '과', '도', '를', '으로', '자', '에', '와', '한', '하다']
more_stops = ['있다','않다','가지','받다','수도','수가', '이다','아니다', '없다', '이기','되다','까지','때문','면서','어요']
more_more = ['위해','대한','가장','정도','이나','어서']
mas = ['입니다','있습니다','같아요','같습니다','있었다','이러한','있으며','대부분','','있는데','따라서','없습니다','여러분','있다는']
stopwords = stopwords + more_stops + more_more + mas


def preprocess(doc, twitter, remove_stopwords = False, stop_words = []):
    
    preprocessed_doc = []

    process = twitter.pos(doc)
    #process = [word for word, tag in process if tag in ['Noun','Verb','Adjective', 'Modifier']]
    process = [word for word, tag in process if tag in ['Noun','Adjective']]

    if remove_stopwords:
        process = [word for word in process if not word in stop_words]
        preprocessed_doc.extend([word for word in process if len(word) > 1])
        
    return ' '.join(preprocessed_doc)
```

```python
%%time
# preprocessing

preprocessed = []
count = 0
docs_len = len(awarn_sentences)

for doc in awarn_sentences:
    preprocessed.append(preprocess(doc, twitter, remove_stopwords=True, stop_words = stopwords))
    count += 1
    if count%200==0:
        print('{:.2f}% complete'.format((count/docs_len)*100))
        
print('complete, {0}\n============='.format(len(preprocessed)))
```

```python
%%time

kwds = []
count = 0
prepro_len = len(preprocessed)

for doc in preprocessed:
    kwds.extend(keywords(doc, words=10).split('\n'))
    
    count += 1
    if count%200==0:
        print('{:.2f}% complete'.format((count/prepro_len)*100))
        
print('complete, len(kwds) : {0}\n============='.format(len(kwds)))
```

# 전처리 -- 9/26

```python
from nltk.tokenize import sent_tokenize
import pandas as pd
import re


def slice_by_sentence(docs): # awarness['all_text']
    sentences = []
    for doc in docs: # each document
        doc = re.sub("[ㄱ-ㅎㅏ-ㅣ]", " ", doc)
        sentences.append(sent_tokenize(doc))
        
    return sentences
```

```python
awarness = pd.read_excel('awarness_blog&cafe.xlsx')#, encoding='utf-8')\
docs = awarness['all_text']
```

```python
sentences = slice_by_sentence(docs)
```

```python
sentences[0]
```

# okt (사전 만들기 전)

```python
from konlpy.tag import Okt
okt = Okt()

stopwords = ['의', '가', '이', '은', '들', '는', '좀', '잘', '걍', '과', '도', '를', '으로', '자', '에', '와', '한', '하다']
more_stops = ['있다','않다','가지','받다','수도','수가', '이다','아니다', '없다', '이기','되다']
stopwords = stopwords + more_stops

def preprocess(doc, okt, remove_stopwords = False, stop_words = []):
    
    preprocessed_doc = []
    for sent in doc:
        process = okt.nouns(sent)

        if remove_stopwords:
            process = [word for word in process if not word in stop_words]
            preprocessed_doc.extend([word for word in process if len(word) > 1])
        
    return preprocessed_doc
```

```python
docs = sentences.copy()
len(docs)
```

```python
# make noun_list
noun_list = []
cnt = 0
docs_len = len(docs)

for doc in docs:
    preprocessed_doc = preprocess(doc, okt, remove_stopwords =True, stop_words=stopwords)
    noun_list.extend(preprocessed_doc)
    
    cnt += 1
    if cnt%200==0:
        print('{:.2f}% complete'.format((cnt/docs_len)*100))
        
print('complete, {0}'.format(len(noun_list)))
```

## count

```python
from collections import Counter

counted = Counter(noun_list)
counted
```

```python
count100 = counted.most_common(100)
```

```python
count100
```

```python
count500 = counted.most_common(500)
```

```python
import re
import pandas as pd
import numpy as np
import csv
```

```python
list_en = []
list_kor = []
cnt = 0
length = len(awarness['all_text'])//2

for row in awarness['all_text'].loc[:5000]:
    try:
        row_kor = re.sub("[^가-힣ㄱ-ㅎㅏ-ㅣ ]", " ", row)
        row_en = re.sub("[^a-zA-Z]", " ", row)
    
    except:
        row_kor = None
        row_en = None

    try:
        list_kor.extend(twitter.pos(row_kor))
        list_en.extend(twitter.pos(row_en))

    except:
        pass
    
    cnt += 1
    #if cnt%200 == 0:
    print("{0:.5f}\%\ complete".format((cnt/length)*100)) 

value_kor, counts_kor = np.unique(np.array(list_kor), return_counts=True)
value_en, counts_en = np.unique(np.array(list_en), return_counts=True)
```

```python
fields = ['value','count']
rows_kor = np.asarray((value_kor, counts_kor)).T
rows_en = np.asarray((value_en, counts_en)).T

with open('sum_kor_first', 'w') as f: 
        
    # using csv.writer method from CSV package 
    write = csv.writer(f) 
        
    write.writerow(fields) 
    write.writerows(rows_kor) 

with open('sum_en_first', 'w') as f: 
        
    # using csv.writer method from CSV package 
    write = csv.writer(f) 
        
    write.writerow(fields) 
    write.writerows(rows_en) 
```

```python
list_en1 = []
list_kor1 = []
cnt = 0

for row in awarness['all_text'].loc[5000:]:
    try:
        row_kor = re.sub("[^가-힣ㄱ-ㅎㅏ-ㅣ ]", " ", row)
        row_en = re.sub("[^a-zA-Z]", " ", row)
    
    except:
        row_kor = None
        row_en = None

    try:
        list_kor1.extend(twitter.pos(row_kor))
        list_en1.extend(twitter.pos(row_en))

    except:
        pass
    
    cnt += 1
    if cnt%200 == 0:
        print("{0:.5f}\%\ complete".format((cnt/length)*100)) 

value_kor1, counts_kor1 = np.unique(np.array(list_kor1), return_counts=True)
value_en1, counts_en1 = np.unique(np.array(list_en1), return_counts=True)
```

```python
fields = ['value','count']
rows_kor1 = np.asarray((value_kor1, counts_kor1)).T
rows_en1 = np.asarray((value_en1, counts_en1)).T

with open('sum_kor_first1', 'w') as f: 
        
    # using csv.writer method from CSV package 
    write = csv.writer(f) 
        
    write.writerow(fields) 
    write.writerows(rows_kor1) 

with open('sum_en_first1', 'w') as f: 
        
    # using csv.writer method from CSV package 
    write = csv.writer(f) 
        
    write.writerow(fields) 
    write.writerows(rows_en1) 
```

# 돌려보자.
전처리는, 노말라이즈, 스템,

okt로 한글 : ['Noun','Verb','Adjective', 'Modifier'], 영어 : ['Alpha']  태그만 사용.
한글은 20개 이상, 영어는 15개 이상으로 

```python
okt = Okt()

posed = okt.pos(awarness['all_text'].iloc[1], norm=True, stem=True)
posed
```

```python
def process_kor_en(posed):
    for word, pos in posed:
#         if pos in ['Noun','Verb','Adjective', 'Modifier']:
#             posed_kor.append((word,pos))
        if pos in ['Alpha']:
            posed_en.append((word.lower(),pos))
```

```python
posed_kor = []
posed_en = []
process_kor_en(posed)
len(posed), len(posed_kor), len(posed_en)
```

```python
posed_en
```

```python
# tag 별로 한글, 영어 나눠 줌
def process_kor_en(posed):
    for word, pos in posed:
#         if pos in ['Noun','Verb','Adjective', 'Modifier']:
#             posed_kor.append((word,pos))
        if pos in ['Alpha']:
            posed_en.append(word.lower())
            


posed_en = []
posed_kor = []
cnt = 0
length = len(awarness['all_text'])
counted_kor = []
counted_en = []



for row in awarness['all_text']:

    try:
        process_kor_en(okt.pos(row, norm=True, stem=True))

    except: #텍스트 없을 경우
        pass
    
    cnt += 1
    if cnt%200 == 0:
        print("{0:.5f}% complete".format((cnt/length)*100)) 

print('100%!!!')
```

```python
from nltk.corpus import stopwords
stops = set(stopwords.words("english"))
stops
```

```python
no_stops_en = [en for en in posed_en if not en in stops]
no_stop_up_1_en = [en for en in no_stops_en if len(en)>1]
len(no_stops_en),len(no_stop_up_1_en)
```

```python
no_stop_up_1_en
```

```python
series_en = pd.Series(no_stop_up_1_en)
vc = series_en.value_counts()
vc
```

```python
working_en = []
for word, cnt in vc.items():
    working_en.append((word, cnt))
working_en
```

# 한글 너무 길어서 포기

```python
test_kor = []
idx = 0
for kor in posed_kor:
    cnt = posed_kor.count(kor)
    if idx%100==0:
        print('{:.2f}'.format((idx/len(posed_kor))*100))
    if (cnt > 100) and ((kor, cnt) not in test_kor):
        test_kor.append((kor, cnt))
    idx += 1
len(test_kor)
```

```python
test_en = []
idx = 0
for en in posed_en:
    cnt = posed_en.count(en)
    if idx%100==0:
        print('{:.2f}'.format((idx/len(posed_en))*100))
    if (cnt > 100) and ((en, cnt) not in test_en):
        test_en.append((en, cnt))
    idx += 1
len(test_en)
```

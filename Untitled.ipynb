```python
import re
import pandas as pd
import numpy as np
import csv
```

```python
list_en = []
list_kor = []
cnt = 0
length = len(awarness['all_text'])//2

for row in awarness['all_text'].loc[:5000]:
    try:
        row_kor = re.sub("[^가-힣ㄱ-ㅎㅏ-ㅣ ]", " ", row)
        row_en = re.sub("[^a-zA-Z]", " ", row)
    
    except:
        row_kor = None
        row_en = None

    try:
        list_kor.extend(twitter.pos(row_kor))
        list_en.extend(twitter.pos(row_en))

    except:
        pass
    
    cnt += 1
    #if cnt%200 == 0:
    print("{0:.5f}\%\ complete".format((cnt/length)*100)) 

value_kor, counts_kor = np.unique(np.array(list_kor), return_counts=True)
value_en, counts_en = np.unique(np.array(list_en), return_counts=True)
```

```python
fields = ['value','count']
rows_kor = np.asarray((value_kor, counts_kor)).T
rows_en = np.asarray((value_en, counts_en)).T

with open('sum_kor_first', 'w') as f: 
        
    # using csv.writer method from CSV package 
    write = csv.writer(f) 
        
    write.writerow(fields) 
    write.writerows(rows_kor) 

with open('sum_en_first', 'w') as f: 
        
    # using csv.writer method from CSV package 
    write = csv.writer(f) 
        
    write.writerow(fields) 
    write.writerows(rows_en) 
```

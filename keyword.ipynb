```python
from nltk.tokenize import sent_tokenize
import pandas as pd
import re
```

```python
awarness = pd.read_excel('awarness_blog&cafe.xlsx')#, encoding='utf-8')\
docs = awarness['all_text']
```

```python
def slice_by_sentence(docs): # awarness['all_text']
    sentences = []
    for doc in docs: # each document
        doc = re.sub("[ㄱ-ㅎㅏ-ㅣ]", " ", doc)
        doc = doc.lower()
        sentences.append(sent_tokenize(doc))
        
    return sentences
```

```python
sentences = slice_by_sentence(docs)
```
